```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python

shirts = pd.read_csv('data/14_shirts.csv', index_col=0)

size_mapping = {"S": 0, "M": 1, "L": 2, "XL": 3}

shirts["taglia"] = shirts["taglia"].map(size_mapping)

shirts = pd.concat([shirts, pd.get_dummies(shirts["colore"])], axis=1).drop('colore', axis=1)

shirts
```
